Classification is a process of dividing the given data points into two or more classes. When the
number of classes is 2, then it is called Binary Classification. In this lab, you will design Deep Neural Network for
Binary Classification that predicts whether a person has a heart disease or not. There are 2 classes:

Heart disease present (target=1)

Heart disease not present (target=0)

**1.Import modules**

```python
import tensorflow as tf
import numpy as np
from sklearn.model_selection import train_test_split
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense
import pandas as pd
from tensorflow.keras.optimizers import RMSprop
import matplotlib.pyplot as plt
```

**2.Load the Dataset**

```python
df = pd.read_csv('heart_data.csv')
```

```python
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.isna().sum().sum()
```

**3.Split the Dataset**

```python
X = df.drop('target',axis=1)
y = df.target
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,train_size=0.80,test_size=0.20)
```

**4.Create a neural network**

```python
model = Sequential()
model.add(Dense(8,input_dim=13,activation='relu'))
model.add(Dense(1,activation='sigmoid'))
```

**5.compile the model**

```python
model.compile(loss='mean_squared_error',metrics=['accuracy'],optimizer=RMSprop(0.001))
```

**6.print the summary**

```python
model.summary()
```

**7.Train the model**

```python
model.fit(X_train,y_train,epochs=200,batch_size=10,verbose=1)
```

**9.evalute the trained model**

```python
model.evaluate(X_test,y_test)
```
